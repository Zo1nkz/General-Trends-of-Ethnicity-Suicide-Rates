<h4> 
    From here, we are going to filter just columns of data we need.
    For the purpose of setting context, the data to be used will be of the ENTIRE population rather than a single ethnicity.
</h4>

```python
df = pd.read_csv('raceAndSuicide.csv', usecols=['YEAR', 'STUB_NAME', 'STUB_LABEL', 'UNIT', 'AGE', 'ESTIMATE'])
df.head(50)
```

<h4> After filtering the columns, we need to filter it by specific data of a certain type/has a specific element. </h4>

```python
df = df[df['STUB_NAME'] == 'Total'].set_index('UNIT')
df.head()
```

<h4> Now that we have our total suicide rates for all people from the years 1950-2018, the last thing we have to do to gather our data is split it between the two units </h4>

```python
dfA = df.filter(like='Deaths per 100,000 resident population, age-adjusted', axis='rows')
dfA.head()
```

```python
dfC = df.filter(like='Deaths per 100,000 resident population, crude', axis='rows')
dfC.head() 
```

<h1> Construct Model(s) & Visuals </h1>

<h4> For the sake of clarity, we are only going to be going in depth with the crude data, as it is more straightforward to handle raw data compared to the AOR data that will take a little bit longer to explain the details surrounding </h4>

```python
dfC = dfC.set_index('YEAR').sort_values(by='YEAR')
dfC.plot(kind = 'line', xlabel = 'Year', ylabel = 'Estimated Suicide Rate')
dfC.head(100)
```
